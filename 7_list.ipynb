### Lists in python


```python
## Project 1 == Create a to do list with list
to_do_list = ["buy groceries", "clean the house", "pay bills", "walk the dog"]

## add some items
to_do_list.append("call mom")
to_do_list.append("finish project")
print("To Do List:", to_do_list)
```

```python
## removing some items
to_do_list.remove("pay bills")
print("Updated To Do List:", to_do_list)
```

```python
## cheking if the task is in list
if "walk the dog" in to_do_list:
    print("Don't forget to walk the dog!")
```

```python
## iterating through a list and print the tasks
print("Here are your tasks:")   
for task in to_do_list:
    print("-", task)
```

```python
## Project 2 == Organizing students grades
grades = [88, 92, 79, 85, 95, 90, 78]

## add to grades
grades.append(84)
grades.append(91)

print("Grades:", grades)
```

```python
## calculate the average of grades
average_grade = sum(grades) / len(grades)
print("Average Grade:", average_grade)
```

```python
## calculate max and min
max_grade = max(grades)
min_grade = min(grades)
print("Highest Grade:", max_grade)
print("Lowest Grade:", min_grade)
```
