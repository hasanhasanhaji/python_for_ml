```python
student = {"name": "Hassan", "age" : 34} # define two key with their values
student
```

```python
## accessing to the elements of a dictionary
print(student["name"])
print(student["age"])
print(student.get("name")) # apprach two
print(student.get("last_name")) 
print(student.get("last_name","Not available")) # if the item does not exist in the dict print error
```

```python
## Modifying Dict items
print(student)
```

```python
student["last_name"] = "Haji mohammadi"
print(student)
```

```python
del student["last_name"] # delete last_name key
print(student)
```

```python
## Dict methods
keys = student.keys()
print(keys)
values = student.values()
print(values)
items = student.items() # get a pair of key and values
print(items)
```

```python
## shallow copy
student_copy = student
print(student)
print(student_copy)
```

```python
student["name"] = "ali"
print(student)
print(student_copy) # this is also change!!!!!!   in next query i will use shallow
```

```python
student_copy2 = student.copy()
print(student)
print(student_copy)
student["name"] = "gholi"
print(student)
print(student_copy2)
```

```python
## Iterating over dicts
## you can iterate over keys,values and items

## iteraitng over keys
for keys in student.keys():
    print(keys)
```

```python
## iterate over values
for values in student.values():
    print(values)
```

```python
## iterate over items
for k,v in student.items():
    print(k,v)
```

```python
## Nested keys
students ={
    "student_1" : {"name":"Hassan", "age" : 25},
    "student_2" : {"name":"Ali", "age" : 33},
    "student_3" : {"name":"Safdar", "age" : 21}
}
students
```
