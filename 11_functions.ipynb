```python
# why function?
num =24
if num % 2 == 0 :
    print("The number is even")
else:
    print("The number is odd!")
```

```python
def even_odd(num):
    """
    This function return if the number is even or odd!
    """
    if num % 2 == 0 :
        print("The number is even")
    else:
        print("The number is odd!")

even_odd(30) # call the function
```

```python
## function with multiple parameters

def add(a,b):
    return a+b

result=add(2,4)
print(result)
    
```

```python
def greet(name="Guest"):
    print(f"Hello {name} Welcome To the paradise")

greet("Hassan")
```

```python
### Variable Length Arguments
## Positional And Keywords arguments

def print_numbers(*nums):
    for number in nums:
        print(number)
print_numbers(1,2,3,4,5,6,7,8,"Hassan")
```

```python
### Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
print_details(name="Hassan",age="32",country="Iran")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional arument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(1,2,3,4,"Krish",name="Hassan",age="32",country="Iran")
```

```python
### Return statements
def multiply(a,b):
    return a*b

multiply(2,3)
```

```python
### Return multiple parameters
def multiply(a,b):
    return a*b,a

multiply(2,3)
```
