## Sets in Python

```python
# Create a set
my_set = {1,2,4,7,9}
print(my_set)
print(type(my_set))
```

```python
empty_set = set()
type( empty_set)
```

```python
new_set = set([1,1,1,2,2])
new_set  ## the duplicate will be removed
```

```python
## Basic set operations
## A dding and removing elements
new_set.add(7)
new_set
```

```python
new_set.remove(7)
new_set
```

```python
new_set.remove(10)
```

```python
new_set.discard(10) # use this does not generate error
new_set
```

```python
# pop method
x = new_set.pop()
print(x)
print(new_set)
```

```python
## create all the elements
new_set.clear()
print(new_set)
```

```python
# Set membership in python
my_set = {1,2,3,4,5}
print(3 in my_set)
```

```python
# Math operations in python
set1 = {1,2,23,5,7}
set2 = {1,2,14,123,22}

union_set = set1.union(set2)
union_set 
```

```python
intersection_set = set1.intersection(set2)
intersection_set
```

```python
set1.difference(set2)
```

```python
set2.difference(set1)
```

```python
set1.symmetric_difference(set2)
```

```python
## Set methods
set1 = {1,2,3}
set2 = {3,4,5}

print(set1.issubset(set2))
print(set1.issuperset(set2))
```

```python
lst =[ 1,2,2,2,2,3,4,5,6,7,7,7,7] 
# how to remove duplicates from the list
set(lst)
```

```python
### Counting unique word in text
text = "Hi iran. god bless iran. god saves iran"
tokens = text.split()
unique_tokens = set(tokens)
print(unique_tokens)
```
