### Loops in Python"

```python
## for loop
for i in range(1,7): # from 1 to 6
    print(i)
    
```

```python
for i in range(4): ## default start is 0
    print(i,"Hello World")
```

```python
for i in range(1,11,2): ## step size of 2
    print(i)
```

```python
for i in range(11,1,-2):    # step size of -2|
    print(i)
```

```python
## strings
str = "Hassan Haji Mohammadi"
for char in str: # iterating through each character of the string
    print(char)
```

```python
## While loop Tuturial
##  the while loop continues to execute as long as the condition is true

count = 0 # initialization
while count < 5: # condition
    print("Count is:", count) # body of the loop
    count += 1  # increment the count to avoid infinite loop    
    
```

```python
## Loop control statements: break, continue, pass
## break statement: terminates the loop completely

count = 0
while True: 
    print("Count is:", count)
    count += 1
    if count >= 5:
        break
```

```python
## continue statement: skips the current iteration and moves to the next iteration
count = 0
while count < 5:
    count += 1
    if count == 3:
        continue  # skip the rest of the loop when count is 3
    print("Count is:", count)
```

```python
# --- IGNORE ---
## pass statement: does nothing, acts as a placeholder
for i in range(5):
    if i == 2:
        pass  # do nothing when i is 2
    else:
        print("i is:", i)
```

```python
## Nested Loops
for i in range(3):  # outer loop
    for j in range(2):  # inner loop
        print(f"i: {i} ----------- j: {j}")
```

```python
## Calculate the sum of first N natural number with while and for

## first with while
n = 15
sum = 0
count = 1
while count <= n:
    sum += count
    count += 1

print(f"The sum of first {n} natural numbers using while loop is: {sum}")
```

```python
## second with for
sum = 0
for i in range(1, n+1):
    sum += i    
print(f"The sum of first {n} natural numbers using for loop is: {sum}")
```

```python
## Example prime numbers between 1 and 100
for num in range(2, 101):
    is_prime = True
    for i in range(2, int(num**0.5) + 1): # check divisibility up to the square root of num
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, "is a prime number")
```
