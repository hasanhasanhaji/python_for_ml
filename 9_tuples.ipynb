## Tuples

```python
## Creating tuple
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
numbers = tuple([1,2,3,4,5,6,7,8])
numbers
```

```python
mixed_tuple = (1,"hassan", True, 3.14)
mixed_tuple
```

```python
numbers[1]
```

```python
numbers[0]
```

```python
numbers[0:2]
```

```python
numbers * 3
```

```python
## Immutable nature of tuples means we can not change the items after creating it
numbers = (1,2,3,4,5,6,7)
numbers[1] = 2
```

```python
## Tuple methods
numbers
```

```python
numbers.count(1) ## search the tuple for the count of input
```

```python
numbers.index(3) ## return the first index that match the value
```

```python
## packing and unpacking

## packing
packed_tuple = "iran" , 1 , 3.14
packed_tuple
```

```python
## unpacking
a,b,c = packed_tuple
print(a)
```

```python
## unpacking with * 
numbers = (1,2,3,4,5,6)
first, * middle, last = numbers
print (first)
print(middle)
print(last)
```

```python
## Nested Tuple and nested list

lst = [[1,2,3], [3,4,"hello"]]
lst[1][2]
```

```python
lst[1][0:2]
```

```python
tple = ((1,2,3), (3,4,"hello"))
tple[0]
```

```python
## iterating over nested tuples
for sub_tuple in tple:
    for item in sub_tuple:
        print(item,end="  ")
    print()
```
